# LSTM原始交易数据不同窗口期对比

## 测试环境

* Intel Core i5-6500 3.2-GHz
* 8.00GB

## 测试内容

```python
import jupyter_helper
from keras.backend import clear_session
import pandas as pd
import numpy as np
```

```python
report={}
report['window']=[]
report['acc']=[]
days=1
```

```python
for i in range(1,11):
    h = jupyter_helper.do(
        window=i,
        days=days,
        wrapper=jupyter_helper.wrapper(),
        norm=jupyter_helper.normalize())
    s=jupyter_helper.show_history(h)
    clear_session()
    report['window'].append(i)
    report['acc'].append(s['score'][2])
```

## 报告

```python
pd.DataFrame(report,columns=['window','acc']).set_index('window').sort_values('acc',ascending=False)
```
