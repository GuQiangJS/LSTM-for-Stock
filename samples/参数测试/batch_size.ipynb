## 测试环境

```python
jupyter_helper.print_system_info()
```

```python
report={}
report['acc']=[]
report['mae']=[]
report['loss']=[]
report['time']=[]
report['utils']=[]
report['window']=[]
report['days']=[]
report['batch_size']=[]
```

```python
times=10
bss=[4096,2048,1024,512,256,128,64,32,16,8,4]
for bs in bss:
    for i in range(times):
        print('{0}/{1}[{2}]'.format(i+1,times,bs))
        ls = [{'type': 'lstm', 'units': 64},
              {'type': 'dense'}]
        h = jupyter_helper.do(layers=ls, summary=False,batch_size=bs)
        s = jupyter_helper.show_history(h, show_plt=False, print_score=False)
        clear_session()
        report['window'].append(h['window'])
        report['days'].append(h['days'])
        report['batch_size'].append(h['batch_size'])
        report['acc'].append(s['score'][2])
        report['mae'].append(s['score'][1])
        report['loss'].append(s['score'][0])
        report['time'].append(h['end']-h['start'])
        report['utils'].append(64)
```
