## 测试环境

```python
jupyter_helper.print_system_info()
```

```python
report={}
report['acc']=[]
report['mae']=[]
report['loss']=[]
report['time']=[]
report['utils']=[]
report['window']=[]
report['days']=[]
report['batch_size']=[]
report['act']=[]
```

```python
times = 10
batch_size = 256
utils = 64
acts = ['softmax', 'elu', 'selu', 'softplus', 'softsign', 'relu',
        'tanh', 'sigmoid', 'hard_sigmoid', 'exponential', 'linear']
for act in acts:
    for i in range(times):
        print('{0}/{1}[{2}]'.format(i+1, times, act))
        ls = [{'type': 'lstm', 'units': utils},
              {'type': 'dense', 'activation': act}]
        h = jupyter_helper.do(layers=ls, summary=False, batch_size=batch_size)
        s = jupyter_helper.show_history(h, show_plt=False)
        clear_session()
        report['window'].append(h['window'])
        report['days'].append(h['days'])
        report['batch_size'].append(h['batch_size'])
        report['acc'].append(s['score'][2])
        report['mae'].append(s['score'][1])
        report['loss'].append(s['score'][0])
        report['time'].append(h['end']-h['start'])
        report['utils'].append(64)
        report['act'].append(act)
```
