## 测试环境

```python
jupyter_helper.print_system_info()
```

```python
report={}
report['acc']=[]
report['mae']=[]
report['loss']=[]
report['time']=[]
report['utils']=[]
report['window']=[]
report['days']=[]
report['batch_size']=[]
```

```python
lst = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512]
for u in lst:
    i_s = [1, 2, 3, 4, 5]
    for i in i_s:
        print('{0}/{1},{2}/{3}'.format(lst.index(u)+1,
                                       len(lst), i_s.index(i)+1, len(i_s)))
        ls = [{'type': 'lstm', 'units': u},
              {'type': 'dense'}]
        h = jupyter_helper.do(layers=ls, summary=False)
        s = jupyter_helper.show_history(h, show_plt=False, print_score=False)
        clear_session()
        report['window'].append(h['window'])
        report['days'].append(h['days'])
        report['batch_size'].append(h['batch_size'])
        report['acc'].append(s['score'][2])
        report['mae'].append(s['score'][1])
        report['loss'].append(s['score'][0])
        report['time'].append(h['end']-h['start'])
        report['utils'].append(u)
```
